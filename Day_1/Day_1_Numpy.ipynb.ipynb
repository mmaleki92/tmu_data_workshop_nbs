# NumPy
The NumPy package provides the "ndarray" object. The NumPy array is used to contain data of uniform type with an arbitrary number of dimensions. NumPy then provides basic mathematical and array methods to lay down the foundation for the entire SciPy ecosystem. The following import statement is the generally accepted convention for NumPy.

```python
import numpy as np
```

## Array Creation
There are several ways to make NumPy arrays. An array has three particular attributes that can be queried: shape, size and the number of dimensions.

```python
a = np.array([1, 2, 3])
print(a.shape)
print(a.size)
print(a.ndim)
```

```python
x = np.arange(100)
print(x.shape)
print(x.size)
print(x.ndim)
```

```python
y = np.random.rand(5, 80)
print(y.shape)
print(y.size)
print(y.ndim)
```

```python
y = np.zeros(80)
print(y.shape)
print(y.size)
print(y.ndim)
```

```python
y = np.ones(80)
print(y.shape)
print(y.size)
print(y.ndim)
```

## Array Manipulation
How to change the shape of an array without a copy!

```python
x.shape = (20, 5)
print(x)
```

NumPy can even automatically figure out the size of at most one dimension for you.

```python
y.shape = (4, 20, -1)
print(y.shape)
```

or

```python
y.reshape(4, 20, -1)
```

## Array Indexing

```python
# Scalar Indexing
print(x[2])
```

```python
# Slicing
print(x[2:5])
```

```python
# Advanced slicing
print("First 5 rows\n", x[:5])
print("Row 18 to the end\n", x[18:])
print("Last 5 rows\n", x[-5:])
print("Reverse the rows\n", x[::-1])
```

```python
# Boolean Indexing
print(x[(x % 2) == 0])
```

```python
# Fancy Indexing -- Note the use of a list, not tuple!
print(x[[1, 3, 8, 9, 2]])
```

## Broadcasting
Broadcasting is a very useful feature of NumPy that will let arrays with differing shapes still be used together. In most cases, broadcasting is faster, and it is more memory efficient than the equivalent full array operation.

```python
print("Shape of X:", x.shape)
print("Shape of Y:", y.shape)
```

```python

arr = np.array([1, 2, 3])
repeated_arr = np.tile(arr, 3)
print(repeated_arr)
```

Another example of broadcasting two 1-D arrays to make a 2-D array.

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-8, 8, 0.25)
print(x.shape, y.shape)
z = x[np.newaxis, :] * y[:, np.newaxis]
print(z.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set the random seed for reproducibility (optional)
np.random.seed(42)

# Generate random numbers from a normal distribution
num_points = 1000
data = np.random.normal(loc=0, scale=1, size=num_points)

# Create a histogram to visualize the data
plt.hist(data, bins=30, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution with Mean = 0, Sigma = 1')

# Display the plot
plt.show()
```

# Functions

```python
a = np.linspace(0, 10, 100)
b = np.exp(a)
plt.plot(a, b)
plt.show()
```

```python
a = np.linspace(0, 10, 100)
b = np.sin(a)
plt.plot(a, b)
plt.show()
```

```python
# create a beautiful plot from an equation of numpy
x = np.linspace(0, 10, 100)
y = np.sin(x) * np.exp(-x)
plt.plot(x, y)
plt.show()
```

```python
# create flower plot from an equation of numpy
theta = np.linspace(0, 2*np.pi, 100)
r = np.sin(5*theta) * np.exp(-0.1*theta)
plt.polar(theta, r)
plt.show()
```

## Data Input and Output	

```python
import numpy as np

# Generate some data
data = np.random.rand(5, 3)
# Write data to a text file
np.savetxt('output.txt', data, delimiter=',')
print("Data saved successfully.")
```

```python
import numpy as np
# Read data from a text file
data = np.loadtxt('output.txt',delimiter=',')
print(data)
```

Reading and writing NumPy binary files

```python
import numpy as np

# Read data from a binary file
data = np.load('data.npy')

print(data)

# Write data to a binary file
np.save('output.npy', data)

print("Data saved successfully.")
```

## Read image

```python
import numpy as np
from PIL import Image

# Load image
image_path = 'cat.jpg'  # Replace with the actual path of your image file
image = Image.open(image_path)

# Convert image to NumPy array
image_array = np.array(image)

# Print array shape and data type
print("Array shape:", image_array.shape)
print("Data type:", image_array.dtype)

# Access pixel values
print("Pixel values at (0, 0):", image_array[0, 0])
```

```python
import matplotlib.pyplot as plt
plt.imshow(image_array)
```
